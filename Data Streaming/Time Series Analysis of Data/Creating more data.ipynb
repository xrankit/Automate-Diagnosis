```python
for z in ['hr','Galvanic skin response','calories','temp'] :
    mean_value=D[z].mean()
    D[z].fillna(value=mean_value, inplace=True)
print(D)
l=['File1.csv','File2.csv','File3.csv','File4.csv','File5.csv']
```

```python
D=D.drop('sec',axis=1)
```

```python
D=D.drop('Time',axis=1)
```

```python
for i in range(5):
    dnew=pd.DataFrame()
    c=random.randrange(2,5)
    dnew['date_time']=D['date_time']
    for j in ['hr','Galvanic skin response','calories','temp']:
        mean_value=D[j].mean()
        arr=np.random.uniform((mean_value-c),(mean_value+c),73440)
        dnew[j]=arr
    dnew.to_csv(l[i])
        
```
